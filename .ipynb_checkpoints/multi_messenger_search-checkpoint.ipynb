## Multi Messenger Detection Using Bayesian Statistics

I am going to be making a simulation for $n$ messenger scenario using Bayesian statistics, with the method explained in the "How to Search for Multiple Messengers—A General Framework Beyond Two Messengers" paper. When we have n=2 messengers, there are 5 sub-hypothesis, which are: both of the messengers being noise; first one being real; second one being real; both of them being real and coming from the same source of astronomical event; both of them being real but coming from different events. For $n$ messengers we have the recursion formula
$$
f(n+1) = \sum_{i=0}^n \begin{pmatrix} n \\ i \end{pmatrix} f(i),\enspace f(0)=1
$$

There is not a uniformly most powerful test for our problem. So what we do is, because we know the astrophysical and noise rate of occurances of these different messengers, we can empirically weight and combine the sub-hypotheses. This reduces the sub-hypotheses from $f(n+1)$ to $f(n)$. Combining more sub-hypotheses requires assuming rates for multi-messenger observations. Due to the small number of such detections, these rates could not be empirically determined and cannot be used in an objective manner. 

The ratio of the predicted number density of multi-messenger sources to the number density of individual messengers’ sources together with sources’ emission models (i.e., emission energies, dependency on inclination etc.) and messengers’ propagations in space, the ratios between the rates of each kind of detection can be found, which is necessary for weighting all of $f (n+1)$ sub-hypotheses. After weighting, null and signal hypotheses reduce to simple hypotheses and the Neyman–Pearson lemma can be used for finding the most powerful test, which gives us the test statistic (TS):
$$
\text{TS}(x) = \frac{P(x|H_s)}{P(x|H_n)} = \frac{\sum_{i} P(x|H_s^i)P(H_s^i)}{\sum_{j} P(x|H_n^j)P(H_n^j)} \times \frac{\sum_{j}P(H_n^j)}{\sum_{i}P(H_s^i)}
$$
$x$ is the complete set of detection outcomes, $H_s$ and $H_n$ are signal and null hypotheses in order, $H_s^i$ and $H_n^j$ are the individual signal and null sub-hypotheses in order. The last term is ignored because it cancels out with the same term that will come from the detection likelihoods.


### Detection Likelihoods $P(x|H_a^b)$

There are two issues to resolve when finding the detection likelihoods. First is to determine if the detection is astrophysical or noise originated. This part can generally be decoupled for different kinds of messengers because of the independence of the detectors. The second is determining the multi-messenger aspect of the detection for which correlations between messengers are required, especially in the spacetime coordinates of the messengers. This coupling can be done with a source model with parameters $\theta$ as
$$
P(x|H_a^b) = \int P(x| \theta, H_a^b)P(\theta | H_a^b) d\theta
$$

The source parameters $\theta$ can include any property of the sources, such as spatial position or emission energies of the sources.

So for example source parameters for multi-messenger detection of high energy neutrinos and gravitational waves will include, $\theta = \{r_s, \Omega_s, t_s, E_{GW}, E_\nu \}$, which are the distance of the source, angular position in the sky, retarded reference time of the event (due to travel time of the messenger), the isotropic equivalent emission energies in GWs, high-energy neutrinos.

It would also be useful to now mention the detection outcomes of the GWs and high-energy neutrinos. The gravitational waves have three detection outcomes $x_{\text{GW}} = \{t_{\text{GW}}, \mathcal{D}, \mathcal{F}\}$, which are: the detection time of the gravitational wave, its joint volume localization and isotropic equivalent emission energy estimation as a four-dimensional probability distribution, and the estimated false alarm rare. 

For high-energy neutrinos, the detection outcomes are $x_\nu = \{t_\nu, \Omega_\nu, \sigma_\nu, \epsilon_\nu \}$, which are: the detection time for neutrinos, the expected sky positions, the angular errors for sky localization, and their reconstructed energies.

If the sub-hypothesis does not require a multi-messenger detection and the detectors are independent from each other, then we can expand the detection likelihoods' probabilities
$$
P(x|H_a^b) = \prod_i P(x_i|H_a^b)
$$
where $i$ runs over all detectors or the individual detections we have. Similarly for when there is common source, we can expand the outcomes' probabilities for a fixed source as a product for different detectors
$$
P(x|\{\theta\}, H_a^b) = \prod_i P(x_i|\{\theta\}, H_a^b).
$$
The notation $\{\theta\} = \{\theta_1, \theta_2, \dots\},$ is used for representing if there are multiple set of sources.

##### Sub-hypothesis Combinations
Another complication arises from the combinatorics if the sub-hypothesis can be satisfied with different groupings of the detections. So if the sub-hypothesis has detection of five particles where the three of them are from a source and rest is noise orignated. In this case, the sub-hypthesis can be satisfied with $\begin{pmatrix} 5 \\ 3 \end{pmatrix} = 10$ combinations. In such cases, we expand the probabilities of $P(x_i | \theta, H_a^b)$ as
$$
P(x_i | \{\theta\}, H_a^b) = \sum_{\{x_i^j,x_i^k, \dots \}} P(x_i| \{\theta\}, H_a^b, \{x_i^j,x_i^k, \dots \}, \{x_i^p, x_i^q, \dots \}) \times P(\{x_i^j,x_i^k, \dots \}, \{x_i^p, x_i^q, \dots \}|\{\theta\}, H_a^b)
$$
although it looks very complicated, this term isn't overly complicated. The sum is over all of the combinations that satisfy the sub-hypothesis $H_a^b$, the sets of $\{x_i^j,x_i^k, \dots \}$ and $\{x_i^p, x_i^q, \dots \}$ are the detection outcomes of individual detections from different sources. So the first braces and second braces are just detections from 2 different sources. The second term in the summation are the inverse of the total possible combinations, so for a single source and $w$ particles emitted from that source, and total of $W$ detections; the probability is simply
$$
P(\{x_i^j,x_i^k, \dots \}, \{x_i^p, x_i^q, \dots \}|\{\theta\}, H_a^b) = \begin{pmatrix} W \\ w \end{pmatrix}^{-1}.
$$

The first term can be found by using physics and empirical data we already have. For memoryless detectors, the detections from different sources are independent so
$$
P(x_i| \{\theta\}, H_a^b, \{x_i^j,x_i^k, \dots \}, \{x_i^p, x_i^q, \dots \}) = P(\{x_i^j,x_i^k, \dots \}| \theta_1, H_a^b)P(\{x_i^p, x_i^q, \dots \}| \theta_2, H_a^b).
$$
These likelihoods can be found via the detector characteristics and emission models. (Idk how that works, ask)

The second term in the integrated probability, $P(\theta | H_a^b)$ can be transformed using the Bayes' rule
$$
P(\theta | H_a^b) = \frac{P(H_a^b |\theta)P(\theta)}{P(H_a^b)}.
$$
The denominator cancels the same term in the test statistic. $P(\theta)$ is the joint density of source parameters being integrated over.

The sub-hypothesis probabilities $P(H_a^b |\theta)$ can be found via the expected counts from the sources or the noise origin by assuming a source density and using the empirically known trigger rate. In simpler words, if the sub-hypothesis $H_a^b$ is from a source, then we can find this probabilty by looking for the expected number of events for such source. If the sub-hypothesis is includes noise, then we can find it by the trigger rate for such source.

```python
def ninty_perc(skymap): # Finding the 90% Credible Prob. Region
    skymap.sort("PROBDENSITY", reverse=True)
    level, ipix = ah.uniq_to_level_ipix(skymap["UNIQ"])
    nside = ah.level_to_nside(level)
    pixel_area = ah.nside_to_pixel_area(nside)

    prob = pixel_area*skymap["PROBDENSITY"]
    cumprob = np.cumsum(prob)
    i = cumprob.searchsorted(0.9)

    area_90 = pixel_area[:i].sum()
    area_90_d2 = area_90.to_value(u.deg**2)
    
    print(area_90, "and ", area_90_d2, "per degrees^2")
    skymap = skymap[:i]
    skymap.sort("UNIQ")
    skymap = skymap["UNIQ",]
    # skymap.write("90percent.moc.fits")
    return skymap
```

```python
def retrieve_event(event_name): # Retreaving the skymap from GraceDB and reading in Qtable format
    gw = client.superevent(event_name)
    gw_dict = gw.json()
    t_GW = gw_dict.get('t_0')
    far = gw_dict.get('far')
    files = client.files(event_name).json()
    if "bayestar.multiorder.fits" in files:
        skymap_url = files["bayestar.multiorder.fits"]
    elif "Bilby.multiorder.fits" in files:
        skymap_url = files["Bilby.multiorder.fits"]
    else:
        multiorder_maps = [s for s in files if s.endswith("multiorder.fits")]
        skymap_url = files[multiorder_maps[0]]
    filename = download_file(skymap_url, cache=True)
    
    skymap = QTable.read(filename)
    return skymap, t_GW, far

skymap, t_GW, far = retrieve_event("S241114y")
print(skymap)
print(t_GW)
print(far)

# Getting Volume Localization

def volume_localization(skymap, distance, right_asc, declination): # Yields the probability of the 3 dimensional point in space: distance and sky location
    uniq = skymap["UNIQ"]
    prob_dens = skymap["PROBDENSITY"]
    distmu = skymap["DISTMU"]
    distsigma = skymap["DISTSIGMA"]
    distnorm = skymap["DISTNORM"]
    
    level, ipix = ah.uniq_to_level_ipix(uniq)
    nside = ah.level_to_nside(level) # Lateralization (from uniq to nested)
    
    ra, dec = right_asc*u.deg, declination*u.deg # Float into degrees
    
    match_ipix = ah.lonlat_to_healpix(ra, dec, nside, order="nested") # longtitude and latitude into
    i = np.flatnonzero(ipix == match_ipix)[0]
    prob_per_deg2 = prob_dens[i].to_value(u.deg**-2) # Total probability of that sky location per degrees^2
    r = distance*u.Mpc
    prob_r = r**2*distnorm[i]*norm(distmu[i], distsigma[i]).pdf(r) # Probability (Gaussian) distribution of distance in the given sky location
    
    prob = prob_per_deg2*prob_r 

    return prob
```

#### Gravitational Wave Likelihoods
##### Signal Likelihood
Gravitational wave signal likelihood can be expanded:
$$
\begin{align}
P(x_\text{GW}|\theta, H_s) & = P(t_{\text{GW}}, \mathcal{D}, \mathcal{F}|r_s, \Omega_s, t_s, E_{GW}, E_\nu , H_s) \\
& = P(t_\text{GW} | t_s, H_s) P(\mathcal{F}|t_\text{GW}, r_s, \Omega_s, E_\text{GW}, H_s) P(\mathcal{D}|t_\text{GW}, \mathcal{F}, r_s, \Omega_s, E_\text{GW}, H_s)
\end{align}
$$
Temportal distribution of $t_\text{GW}$ is assumed to be uniform around $t_s$: $P(t_\text{GW}|t_s, H_s) = (t_\text{GW}^+ - t_\text{GW}^-)^{-1}$ for $t_\text{GW} - t_s = [t_\text{GW}^-, t_\text{GW}^+]$ and $0$ otherwise. Take $t_\text{GW}^+ = - t_\text{GW}^- = 250s$.

Likelihood for volume localization can be expanded using the Bayes' rule:
$$
P(\mathcal{D}|t_\text{GW}, \mathcal{F}, r_s, \Omega_s, E_\text{GW}, H_s) = \frac{P(r_s, \Omega_s, E_\text{GW} | t_\text{GW}, \mathcal{D}, \mathcal{F}, H_s) P(\mathcal{D}|t_\text{GW}, \mathcal{F}, H_s)}{P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, \mathcal{F}, H_s)}
$$
first term in the numertor is the $\mathcal{D}$ distribution itself. Again use Bayes' rule for the denominator:
$$
P(\mathcal{D}|t_\text{GW}, \mathcal{F}, r_s, \Omega_s, E_\text{GW}, H_s) = \mathcal{D}(r_s, \Omega_s, E_\text{GW}) \frac{P(\mathcal{F}|t_\text{GW}, H_s)}{P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s) P(\mathcal{F}|t_\text{GW}, r_s, \Omega_s, E_\text{GW}, H_s)} \times {P(\mathcal{D}|t_\text{GW}, \mathcal{F}, H_s)}.
$$
The last term is disregarded because we assume its independent on the hypothesis and same term is in the null hypothesis and cancels out. Also $P(\mathcal{F}|t_\text{GW}, r_s, \Omega_s, E_\text{GW}, H_s)$'s cancel out as well. The term in the numerator, $P(\mathcal{F}|t_\text{GW}, H_s)$ can be computed by integrating the likelihood for fixed source parameters (which can be obtained from calculations or simulations) over the source parameter:
$$
P(\mathcal{F}|t_\text{GW}, H_s) = \int P(\mathcal{F}|t_\text{GW}, r_s, \Omega_s, E_\text{GW}, H_s) P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s) dr_s d\Omega_s dE_\text{GW}.
$$
So omitting the cancelled out terms we have:
$$
\begin{align}
P(x_\text{GW}|\theta, H_s) & = P(t_\text{GW} | t_s, H_s) P(\mathcal{F}|t_\text{GW}, r_s, \Omega_s, E_\text{GW}, H_s) P(\mathcal{D}|t_\text{GW}, \mathcal{F}, r_s, \Omega_s, E_\text{GW}, H_s) \\
& = P(t_\text{GW} | t_s, H_s) \mathcal{D}(r_s, \Omega_s, E_\text{GW}) \frac{P(\mathcal{F}|t_\text{GW}, H_s)}{P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s)}
\end{align}
$$
We can find the source parameter distribution in the denominator 
$$
P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s) = \frac{\frac{r_s^2}{4\pi} \big[\rho(\frac{E_\text{GW}}{r_s^2}, \Omega_s, t_\text{GW}) \geq \rho_\text{th} \big]}{(1+z(r_s))^4 E_\text{GW} N_r' \log(100)}
$$

##### Null Likelihood
Similarly for the null hypothesis, we can expand the likelihood as
$$
P(x_\text{GW}|H_n) = P(t_\text{GW}|H_n) P(\mathcal{F}|t_\text{GW}, H_n) {P(\mathcal{D}|t_\text{GW}, \mathcal{F}, H_n)} 
$$
the last term is cancelled with the same term in signal likehood.
$$
P(t_\text{GW}|H_n) = T_\text{obs}^{-1}
$$
uniform distrubtion in the observation period. Lastly we can find the second term $P(\mathcal{F}|t_\text{GW}, H_n)$ empirically through the unphysical time shifted coincidences. (I looked at LIGO's false alarm rates, so they have two classifications of false alarm rates. One for low-significance which has $\mathcal{F}_\text{low} = 2.3 \times 10^{-5}$ Hz (2 per day), and significant which has $\mathcal{F}_\text{significant} = 6.34 \times 10^{-8}$ Hz (once per six months)).


To make it simpler, lets write here only the things that needed to be coded. Firstly for the gravitational wave signal hypothesis, we need four distributions: temporal distribution
$$
P(t_\text{GW}|t_s, H_s) = \begin{cases} (t_\text{GW}^+ - t_\text{GW}^-)^{-1} & t_\text{GW} - t_s \in [t_\text{GW}^-, t_\text{GW}^+] \\
                                        0 & \text{otherwise}
\end{cases}
$$
for joint volume localization:
$$
\mathcal{D}(r_s, \Omega_s, E_\text{GW}) = \mathcal{D}(r_s, \Omega_s)
$$
its energy independent here so it becomes the three dimensional probability distribution which we can directly obtain by skymaps.

For the last term we have
$$
\frac{P(\mathcal{F}|t_\text{GW}, H_s)}{P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s)}
$$
where
$$
P(\mathcal{F}|t_\text{GW}, H_s) = \int P(\mathcal{F}|t_\text{GW}, r_s, \Omega_s, E_\text{GW}, H_s) P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s) dr_s d\Omega_s dE_\text{GW}
$$
and
$$
P(r_s, \Omega_s, E_\text{GW}|t_\text{GW}, H_s) = \frac{\frac{r_s^2}{4\pi} \big[\rho(\frac{E_\text{GW}}{r_s^2}, \Omega_s, t_\text{GW}) \geq \rho_\text{th} \big]}{(1+z(r_s))^4 E_\text{GW} N_r' \log(100)}
$$
We replace the $E_\text{GW}$s with the masses of the colliding objects $M_1$ and $M_2$ and instead of SNR (signal-to-noise ratio) function and cut off, we will use the false alarm rates. The redshift too will be ignored because of its insignificance and it should be time independent. So what we get is:
$$
P(r_s, \Omega_s, M_1, M_2| H_s) = \frac{\frac{r_s^2}{4\pi} \big[\mathcal{F}(r_s, \Omega_s, M_1, M_2) \leq \mathcal{F}_\text{th} \big]}{ M_1 M_2 N_r'}
$$
and
$$
P(\mathcal{F}|t_\text{GW}, H_s) = \int P(\mathcal{F}|r_s, \Omega_s, M_1, M_2, H_s) P(r_s, \Omega_s, M_1, M_2| H_s) dr_s d\Omega_s dM_1 dM_2
$$
So some terms and how we can obtain them are in order to be explained. First, $\mathcal{F}_\text{th} = 2.3 \times 10^{-5}$ Hz or (2 per day). Other terms we can pull from the simulated data set "GWTC-3: Compact Binary Coalescences Observed by LIGO and Virgo During the Second Part of the Third Observing Run — O3 search sensitivity estimates"

```python
# Reading the O3 search sensitivity estimates simulation data

injection_file = "endo3_bnspop-LIGO-T2100113-v12.hdf5"
# injection_file = "endo3_bbhpop-LIGO-T2100113-v12.hdf5"

with h5py.File(injection_file, 'r') as f:
    N_draw = f.attrs["total_generated"]
    
    gpstime_source = f['injections/gps_time'][:] # s
    mass1_source = f['injections/mass1_source'][:] # M_solar
    mass2_source = f['injections/mass2_source'][:] # M_solar
    distance_source = f['injections/distance'][:] # Mpc
    inc_source = f['injections/inclination'][:]*180/np.pi # radians to degrees (-90, 90)
    dec_source = f['injections/declination'][:]*180/np.pi # radians to degrees (-90, 90)
    ra_source = f['injections/right_ascension'][:]*180/np.pi # radians to degrees (0,360)
    # far_cwb = f['injections/far_cwb'][:] # Off for the bns systems
    far_gstlal = f['injections/far_gstlal'][:]/365.5 # per day
    far_mbta = f['injections/far_mbta'][:]/365.5 # per day
    far_pycbc_hyperbank = f['injections/far_pycbc_hyperbank'][:]/365.5 # per day
    far_pycbc_bbh = f['injections/far_pycbc_bbh'][:]/365.5 # per day


gpstime_source = np.array(gpstime_source)
mass1_source = np.array(mass1_source)
mass2_source = np.array(mass2_source)
distance_source = np.array(distance_source)
inc_source = np.array(inc_source)
ra_source = np.array(ra_source)
dec_source = np.array(dec_source)
far_gstlal = np.array(far_gstlal)
far_mbta = np.array(far_mbta)
far_pycbc_hyperbank = np.array(far_pycbc_hyperbank)
far_pycbc_bbh = np.array(far_pycbc_bbh)

alt_source = np.concatenate((np.load('alt_array0.npy'), np.load('alt_array1.npy'),
                              np.load('alt_array2.npy'), np.load('alt_array3.npy')))
az_source =  np.concatenate((np.load('az_array0.npy'), np.load('az_array1.npy'),
                              np.load('az_array2.npy'), np.load('az_array3.npy')))

location = EarthLocation(lon=0*u.deg, lat=0*u.deg, height=0*u.m)
```

```python
def P_signal_GW_t(t_GW, t_s): # Signal likelihood temporal distribution function
    t_plus = 250
    t_minus = -250
    if t_minus <= t_GW - t_s <= t_plus:
        return (t_plus - t_minus)**-1
    else:
        return 0


def cut_off(far): # Cut off function of the false alarm, eliminates the low-significance events
    if far <= 2: # 2 per day
        return 1
    else:
        return 0

    
def match_far(gpstime, distance, right_ascension, declination, mass1, mass2):
    t = Time(gpstime*u.s, format='gps')
    altaz = AltAz(obstime = t, location = location)
    observation = SkyCoord(ra = right_ascension*u.deg, dec = declination*u.deg, frame='icrs')
    observation_alt_az = observation.transform_to(altaz)
    alt = observation_alt_az.alt.value
    az = observation_alt_az.az.value

    cos_angular_distance = np.sin(alt_source)*np.sin(alt)+np.cos(alt_source)*np.cos(alt)*np.cos(az_source-az)
    cos_angular_distance[cos_angular_distance > 1] = 1
    cos_angular_distance[cos_angular_distance < -1] = -1
    angular_distance = np.arccos(cos_angular_distance)*180/np.pi

    deviation_sky_location = angular_distance**2/((5)**2)
    # deviation_inc = (np.cos(inc_source) - np.cos(inclination))**2/0.1**2
    deviation_distance = (distance_source - distance)**2/(distance/10)**2
    deviation_mass = ((mass1_source - mass1) + (mass2_source - mass2))**2/((mass1 + mass2)*0.2)**2

    total_deviation = deviation_sky_location + deviation_distance + deviation_mass

    count = 0
    deviation_sorted = np.argsort(total_deviation)
    index = deviation_sorted[count]
    while (far_gstlal[index] >= 2 or far_mbta[index] >= 2):
        count += 1
        if count == len(gpstime_source):
            return np.inf
        else:
            index = deviation_sorted[count]
    return np.min([(far_gstlal[index] <= 2)*far_gstlal[index], (far_mbta[index] <= 2)*far_mbta[index]])


def far_list(gpstime, distance, right_ascension, declination, mass1, mass2):
    t = Time(gpstime*u.s, format='gps')
    altaz = AltAz(obstime = t, location = location)
    observation = SkyCoord(ra = right_ascension*u.deg, dec = declination*u.deg, frame='icrs')
    observation_alt_az = observation.transform_to(altaz)
    alt = observation_alt_az.alt.value
    az = observation_alt_az.az.value

    cos_angular_distance = np.sin(alt_source)*np.sin(alt)+np.cos(alt_source)*np.cos(alt)*np.cos(az_source-az)
    cos_angular_distance[cos_angular_distance > 1] = 1
    cos_angular_distance[cos_angular_distance < -1] = -1
    angular_distance = np.arccos(cos_angular_distance)*180/np.pi
    
    sky_location_condition = (angular_distance <= 15.0)
    distance_condition = ((distance_source - distance)**2 <= (distance/5)**2)
    mass_condition = (mass1*0.65 <= mass1_source) & (mass1_source <= mass1*1.35) & \
                        (mass2*0.65 <= mass2_source) & (mass2_source <= mass2*1.35)

    total_condition = sky_location_condition & distance_condition &  mass_condition
    
    pre_far_list = np.concatenate((far_gstlal[total_condition], far_mbta[total_condition], far_pycbc_hyperbank[total_condition]))
    far_list = pre_far_list[pre_far_list <= 2]
 
    return far_list


def P_F(false_alarm_rate, gpstime, distance, right_ascension, declination, mass1, mass2): # P(F|source parameters)
    fars = far_list(gpstime, distance, right_ascension, declination, mass1, mass2)
    if len(fars) == 0:
        return 0
    else:
        deviation_far = (false_alarm_rate - fars)**2/(false_alarm_rate/10)**2
        count = 0
        for deviation in deviation_far:
            if deviation <= 5:
                count += 1
        return count/len(fars)
    
    
def P_signal_GW_source(gpstime, distance, ra, dec, M_1, M_2): # source parameter distribution for gravitational wave
    far = match_far(gpstime, distance, ra, dec, M_1, M_2)
    r_max = max(distance_source)
    normalization_r = r_max**3/3 # normalizing the distance distribution
    m1_max, m1_min = np.max(mass1_source), np.min(mass1_source)
    m2_max, m2_min = np.max(mass2_source), np.min(mass2_source)
    normalization_mass1, normalization_mass2 = np.log(m1_max/m1_min), np.log(m2_max/m2_min) # normalization of the mass distributions
    nominator = distance**2/(4*np.pi)*cut_off(far)
    denominator = M_1*M_2*normalization_r*normalization_mass1*normalization_mass2
    return nominator/denominator


def P_signal_GW_F(false_alarm_rate): # Monte Carlo Integration over the source parameters
    integral_sum = 0
    far_mask = np.logical_or(far_gstlal <= 2, far_mbta <= 2)
    gpstime_filtered = gpstime_source[far_mask]
    distance_filtered = distance_source[far_mask]
    ra_filtered = ra_source[far_mask]
    dec_filtered = dec_source[far_mask]
    mass1_filtered = mass1_source[far_mask]
    mass2_filtered = mass2_source[far_mask]
    N = len(gpstime_filtered)
    for i in range(N//25):
        gpstime = gpstime_filtered[i]
        r_s = distance_filtered[i]
        ra = ra_filtered[i]
        dec = dec_filtered[i]
        M1 = mass1_filtered[i]
        M2 = mass2_filtered[i]
        
        p_far = P_F(false_alarm_rate, gpstime, r_s, ra, dec, M1, M2)
        if p_far == 0:
            p_source = 0
        else:
            p_source = P_signal_GW_source(gpstime, r_s, ra, dec, M1, M2)
        
        # Multiply the likelihood and prior and add to the sum
        integral_sum += p_far * p_source
    
    # Average the sum to get the Monte Carlo estimate of the integral
    integral_value = integral_sum/(N//25)
    return integral_value


def P_null_GW_t(t_GW, t_start, t_end):
    if t_start <= t_GW <= t_end:
        T_obs = t_end - t_start
        return T_obs**-1
    else:
        return 0
    
def P_null_GW_F(significance):
    if significance == 'low-significance':
        return 2.3*10**-5 # Hz or 2 per day
    if significance == 'significant':
        return 6.34*10**-8 # Hz or once per six months
```

#### High-Energy Neutrino Likelihoods

##### Signal Likelihood
Signal likelihoods for neutrinos can be expanded:
$$
P(t_\nu, \epsilon_\nu, \sigma_\nu, \Omega_\nu|\theta, H_s) = P(t_\nu|t_s, H_s) P(\epsilon_\nu|\Omega_s, H_s) P(\Omega_\nu, \sigma_\nu|\epsilon_\nu, \Omega_s, H_s)
$$
Similarly for neutrinos, the temporal distrubition is assumed to be uniform around $t_s$: $P(t_\nu|t_s, H_s) = (t_\nu^+ - t_\nu^-)^{-1}$ for $t_\nu - t_s = [t_\nu^-, t_\nu^+]$ and $0$ otherwise. Take $t_\nu^+ = - t_\nu^- = 250s$.

The estimated source localization from the detection can be written as a Gaussian distribution 
$$
P(\Omega_s|\epsilon_\nu, \sigma_\nu, \Omega_\nu, H_s) = \frac{e^{\frac{-|\Omega_\nu - \Omega_s|^2}{2\sigma_\nu^2}}}{2\pi \sigma_\nu^2}
$$
We need the probability for the $P(\Omega_\nu, \sigma_\nu|\epsilon_\nu, \Omega_s, H_s)$ so we expand it using the Bayes' rule 
$$
\begin{align}
P(\Omega_\nu, \sigma_\nu|\epsilon_\nu, \Omega_s, H_s) & = \frac{P(\Omega_s|\epsilon_\nu, \sigma_\nu, \Omega_\nu, H_s) P(\sigma_\nu, \Omega_\nu|\epsilon_\nu, H_s)}{P(\Omega_s|\epsilon_\nu, H_s)} \\
& = \frac{e^{\frac{-|\Omega_\nu - \Omega_s|^2}{2\sigma_\nu^2}}}{2\pi \sigma_\nu^2} \frac{P(\sigma_\nu, \Omega_\nu|\epsilon_\nu, H_s)}{P(\Omega_s|\epsilon_\nu, H_s)} \\
& = \frac{e^{\frac{-|\Omega_\nu - \Omega_s|^2}{2\sigma_\nu^2}}}{2\pi \sigma_\nu^2} \frac{P(\epsilon_\nu|H_s)}{P(\epsilon_\nu|\Omega_s, H_s) P(\Omega_s|H_s)} \times P(\sigma_\nu, \Omega_\nu|\epsilon_\nu, H_s)
\end{align}
$$
we used Bayes' rule again to expand the denominator in last line. The last term is not dependent on the hypothesis so it cancels out with the same term in the null likelihood. $P(\epsilon_\nu|\Omega_s, H_s)$ cancels with the second term in the signal likelihood. We assume power law with exponent -2 for the energy distribution of neutrinos and by using the effective area of the neutrino detector $A_{\text{eff}(\epsilon_\nu, \Omega_s)}$, then we have 
$$
P(\epsilon_\nu|H_s) = \frac{\int A_\text{eff}(\epsilon_\nu, \Omega_s) \epsilon_\nu^{-2} P(\Omega_s|H_s) d\Omega_s}{\int_{\epsilon_\text{min}}^{\epsilon_\text{max}} \int A_\text{eff}(\epsilon_\nu', \Omega_s) \epsilon_\nu'^{-2} P(\Omega_s|H_s) d\Omega_s d\epsilon'_\nu}
$$
$\epsilon_\text{min}, \enspace \epsilon_\text{max}$ are 100 GeV and 100 PeV for IceCube. We can find the $A_\text{eff}(\epsilon_\nu, \Omega_s)$ from IceCube's releases (FIND IT).

The source parameter distribution in the last equation is 
$$
P(\Omega_s|H_s) = \frac{\int_{\epsilon_\text{min}}^{\epsilon_\text{max}} A_\text{eff}(\epsilon_\nu, \Omega_s) \epsilon_\nu^{-2} d\epsilon_\nu}{\int \int_{\epsilon_\text{min}}^{\epsilon_\text{max}} A_\text{eff}(\epsilon_\nu, \Omega_s') \epsilon_\nu^{-2} d\epsilon_\nu d\Omega'_s}
$$

##### Null likelihood
Similarly for null likelihood, we expand 
$$
P(t_\nu, \epsilon_\nu, \sigma_\nu, \Omega_\nu|H_n) = P(t_\nu|H_n) P(\epsilon_\nu|t_\nu, H_n) P(\sigma_\nu, \Omega_\nu|\epsilon_\nu\, t_\nu, H_n)
$$
$P(t_\nu|H_n) = T_\text{obs}^{-1}$ and $P(\epsilon_\nu|t_\nu, H_n)$ can be found empirically from detector characteristics and past observations. Last term can be expanded as 
$$
P(\sigma_\nu, \Omega_\nu|\epsilon_\nu\ t_\nu, H_n) = P(\sigma_\nu, \Omega_\nu|\epsilon_\nu\, H_n) \mathcal{T}(t_\nu, \epsilon_\nu, \Omega_\nu)
$$
first term is independent of hypothesis so cancels out with the same term in signal likelihood. Second term comes from the annual modulation due to Earth's motion around the Sun, whose average over one year for every $(\epsilon_\nu, \Omega_\nu)$ pair is one.
